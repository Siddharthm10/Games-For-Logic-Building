```python
#Astrologers stars
f = open("Stars.txt",'w')
inp = int(input("Enter the no. of lines to be printed"))

check = input("Order y-> ascending or n->descending")

if(check == 'y'):
    for i in range(0, inp):
        f.write("*"*(i+1))
        f.write("\n")
elif(check == 'n'):
    for i in range(0,inp):
        f.write("*"*(inp-i))
        f.write("\n")
f.write("Stars")
f.close()
```

```python
import datetime
import pygame 

from pygame import mixer 
  
# Starting the mixer 
mixer.init() 
  
# Loading the song 
mixer.music.load("Iktara.mp3") 
  
# Setting the volume 
mixer.music.set_volume(0.7) 
  
# Start playing the song 
mixer.music.play() 
  
# infinite loop 
while True: 
      
    print("Press 'p' to pause, 'r' to resume") 
    print("Press 'e' to exit the program") 
    query = input("  ") 
      
    if query == 'p': 
  
        # Pausing the music 
        mixer.music.pause()      
    elif query == 'r': 
  
        # Resuming the music 
        mixer.music.unpause() 
    elif query == 'e': 
  
        # Stop the mixer 
        mixer.music.stop() 
        break

```
